## Part 4 - Jittering with stripplot（抖动图）
Usually, multiple data points have exactly the same X and Y values.  As a result, multiple point draws overlap and are hidden.  To avoid this, shake the data points a little so that you can see them visually.  

### Step 1: Import dataset & check dataset size

```python
# Import Data
df = pd.read_csv("https://raw.githubusercontent.com/bdfd/Jupyter_Notebook_PlayGround/main/dataset%20collection/mpg_ggplot2.csv")

# Draw Stripplot
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)    
sns.stripplot(df.cty, df.hwy, jitter=0.25, size=8, ax=ax, linewidth=.5)

# Decorations
plt.title('Use jittered plots to avoid overlapping of points', fontsize=22)
plt.show()
```
