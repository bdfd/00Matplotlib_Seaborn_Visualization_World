## Part 9 - Pairwise Plot（矩阵图）

Pairwise plot are a favorite in exploratory analysis and are used to understand the relationships between all possible pairs of numerical variables.  It is a necessary tool for bivariate analysis.  

### Step 1: Import dataset

```python
# Load Dataset
df = sns.load_dataset('iris')

# Plot
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="scatter", hue="species", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```

```python
# Load Dataset
df = sns.load_dataset('iris')

# Plot
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="reg", hue="species")
plt.show()
```
